### Getting started With Langchain And Open Al

#### In this quickstart we'll see how to:
* Get setup with LangChain, LangSmith and LangServe
* Use the most basic and common components of LangChain: prompt templates, models, and output parsers.
* Build a simple application with LangChain
* Trace your application with LangSmith
* Serve your application with LangServe

```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ['GROQ_API_KEY']=os.getenv("GROQ_API_KEY")

## langsmith Tracking
os.environ["LANGCHAIN_TRACING_V2"] = "true"
os.environ["LANGSMITH_API_KEY"] = os.getenv("LANGSMITH_API_KEY")
os.environ["LANGCHAIN_PROJECT"] = os.getenv("LANGCHAIN_PROJECT")
```

```python
from langchain_groq import ChatGroq
llm = ChatGroq(model="llama-3.1-8b-instant")
print(llm)
```

```python
### Input and get response from LLM

result = llm.invoke("Who is quentin tarantino?")
```

```python
print(result)
print(result.content)
```

```python
### Chat Prompt Template - how we want llm to behave/to act/what kind of a response/role 

from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages(
    [
        ("system","You are an expert AI Engineer.Provide me answers based on the question"),
        ("user","{input}")
    ]
)

prompt
```

```python
## Withour Parser
chain = prompt | llm

response = chain.invoke({"input":"Who is Christopher Nolan"})
print(response)

print(type(response))
```

```python
from langchain_core.output_parsers import StrOutputParser

output_parser = StrOutputParser()
```

```python
## With output_parser

chain = prompt | llm | output_parser

response_new = chain.invoke({"input":"Who is Christopher Nolan"})
print(response_new)

print(type(response_new))
```
