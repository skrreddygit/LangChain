```python
from langchain_community.document_loaders import TextLoader
loader=TextLoader('speech.txt')
docs=loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size = 500,chunk_overlap=50)
final_documents = text_splitter.split_documents(docs)  ## create_documents work only with text files
final_documents
```

```python
### Vector Embedding and Vector Store(ChromaDB)

from langchain_community.vectorstores import Chroma

vector_store = Chroma.from_documents(final_documents,embeddings=1024)
vector_store
```

```python
## how to query from our Chromadb

## Similarity Search

## Retrieve the results from query vectorstore db

query="we refuse to believe that the bank of justice is bankrupt"
retrieved_result = vector_store.similarity_search(query)
retrieved_result
```
